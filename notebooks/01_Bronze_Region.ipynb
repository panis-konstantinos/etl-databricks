# Convert .CSV file to a Delta Table (Region) 

```python
from pyspark.sql.functions import current_timestamp # Import current_timestamp function

# Read Region CSV (tab separated)
df = spark.read.option("header", True).option("sep", "\t").option("inferSchema", True).csv("/Volumes/workspace/default/landing/Region.csv")

# Clean column names
df = df.toDF(*[c.strip().lower().replace(" ", "_") for c in df.columns])

# Add ingestion timestamp
df = df.withColumn("ingestion_timestamp", current_timestamp())

# Save as Bronze table
df.write.mode("overwrite").format("delta").saveAsTable("workspace.default.bronze_region")

# Preview
display(df)
```

# Bronze Layer Validation – Table Verification

```python
%sql
SELECT *
FROM system.information_schema.tables
WHERE table_schema = 'default'
AND table_name LIKE 'bronze%'
```

```python
%sql
DROP TABLE IF EXISTS workspace.default.bronze_region;
DROP TABLE IF EXISTS workspace.default.bronze_product;
DROP TABLE IF EXISTS workspace.default.bronze_reseller;
DROP TABLE IF EXISTS workspace.default.bronze_salesperson;
DROP TABLE IF EXISTS workspace.default.bronze_salesperson_region;
DROP TABLE IF EXISTS workspace.default.bronze_targets;
DROP TABLE IF EXISTS workspace.default.bronze_sales_2017;
DROP TABLE IF EXISTS workspace.default.bronze_sales_2018;
DROP TABLE IF EXISTS workspace.default.bronze_sales_2019;
DROP TABLE IF EXISTS workspace.default.bronze_sales_2020;
```

```python
from pyspark.sql.functions import current_timestamp

BASE = "/Volumes/workspace/default/landing"

def ingest_tab_csv(file_name: str, table_name: str):
    path = f"{BASE}/{file_name}"

    df = (spark.read.option("header", True).option("sep", "\t").option("inferSchema", True)
          .csv(path))

    # Standardized column normalization (Bronze layer naming convention)
    df = df.toDF(*[c.strip().lower().replace(" ", "_") for c in df.columns])

    # Add timestamp (one extra column)
    df = df.withColumn("ingestion_timestamp", current_timestamp())

    # Save as Bronze Delta table
    (df.write
      .mode("overwrite")
      .option("overwriteSchema", "true")
      .format("delta")
      .saveAsTable(f"workspace.default.{table_name}"))

    print(f"Created {table_name} | cols={len(df.columns)}")

jobs = [
    ("Region.csv",            "bronze_region"),
    ("SalespersonRegion.csv", "bronze_salesperson_region"),
    ("Salesperson.csv",       "bronze_salesperson"),
    ("Reseller.csv",          "bronze_reseller"),
    ("Product.csv",           "bronze_product"),
    ("Targets.csv",           "bronze_targets"),

    ("Sales_2017.csv",        "bronze_sales_2017"),
    ("Sales_2018.csv",        "bronze_sales_2018"),
    ("Sales_2019.csv",        "bronze_sales_2019"),
    ("Sales_2020.csv",        "bronze_sales_2020"),
]

for file_name, table_name in jobs:
    ingest_tab_csv(file_name, table_name)
```
