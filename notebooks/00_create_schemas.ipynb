## Overview
This notebook setups the working environment. Creates a catalog, catalog's schemas and a volume in bronze schema to upload the raw data files.

```python
# Create project catalog
spark.sql("CREATE CATALOG IF NOT EXISTS project")
print("Created 'project' catalog.")

# Create schemas (one for each medallion)
spark.sql("CREATE SCHEMA IF NOT EXISTS project.bronze")
spark.sql("CREATE SCHEMA IF NOT EXISTS project.silver")
spark.sql("CREATE SCHEMA IF NOT EXISTS project.gold")
print("Created schemas.")

# Create volume in bronze schema to store raw data
spark.sql("CREATE VOLUME IF NOT EXISTS project.bronze.raw_data")
print("Created volume in bronze schema.")
```
