# Overview
This notebook reverts workspace in initial state to re-run notebooks (or job). There are two options:
- **Option 1**: Delete all catalog and its contents
- **Option 2**: Delete only catalog's schemas and tables

## Option 1
Delete all catalog with its contents.

**WANRING**: This will delete all schemas, tables, volumes etc in the catalog. Raw data should be uploaded again after deletion.

```python
# Delete catalog with all its contents
spark.sql("DROP CATALOG project CASCADE")
```

## Option 2
Delete all medallion schemas and tables in project catalog.

**NOTE**: This will delete only the medallion schemas and tables in 'project' catalog. The volumes that stores raw data will **NOT** be deleted. 

```python
# Define schemas
schemas = ["project.bronze", "project.silver", "project.gold"]

# Drop all tables in medallion schemas
for schema in schemas:
    # List schema tables
    tables = spark.catalog.listTables(schema)

    # Drop tables sequentially
    for table in tables:
        spark.sql(f"DROP TABLE IF EXISTS {schema}.{table.name}")

    print(f"All tables dropped from {schema}")
```
