```python
schema = "project.gold"
tables = spark.catalog.listTables(schema)


for table in tables:
  table_name = f"{schema}.{table.name}"
  display_cnt = f"{table.name}_total_rows"
  display(spark.sql(f"SELECT COUNT(*) AS {display_cnt} FROM {table_name}"))
```
